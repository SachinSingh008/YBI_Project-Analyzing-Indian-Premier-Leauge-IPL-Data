```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
data = {
    'season': [random.randint(2017, 2023) for _ in range(105)],
    'team1': [random.choice(['MI', 'CSK', 'RCB', 'KKR', 'DC', 'SRH', 'RR', 'PBKS', 'GT', 'LSG']) for _ in range(105)],
    'team2': [random.choice(['MI', 'CSK', 'RCB', 'KKR', 'DC', 'SRH', 'RR', 'PBKS', 'GT', 'LSG']) for _ in range(105)],
    'winner': [random.choice(['MI', 'CSK', 'RCB', 'KKR', 'DC', 'SRH', 'RR', 'PBKS', 'GT', 'LSG']) for _ in range(105)],
    'city': [random.choice(['Mumbai', 'Chennai', 'Bangalore', 'Kolkata', 'Delhi', 'Hyderabad', 'Jaipur', 'Mohali', 'Ahmedabad', 'Lucknow']) for _ in range(105)],
    'player_of_match': [random.choice(['Rohit Sharma', 'MS Dhoni', 'Virat Kohli', 'Andre Russell', 'Rishabh Pant', 'David Warner', 'Sanju Samson', 'KL Rahul', 'Shubman Gill', 'Marcus Stoinis', 'Kieron Pollard', 'AB de Villiers', 'Ravindra Jadeja', 'Sunil Narine', 'Shikhar Dhawan']) for _ in range(105)],
    'runs_scored': [random.randint(120, 250) for _ in range(105)],
}

ipl_df = pd.DataFrame(data)

print(ipl_df.head())
```

```python
print(ipl_df.head())
print(ipl_df.info())  
```

```python
team_wins = ipl_df['winner'].value_counts()
print("\nTeam Wins:\n", team_wins)
```

```python
player_of_match_counts = ipl_df['player_of_match'].value_counts()
print("\nPlayer of the Match Counts:\n", player_of_match_counts)
```

```python
average_runs_per_match = ipl_df['runs_scored'].mean()
print("\nAverage Runs Scored per Match:", average_runs_per_match)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='winner', data=ipl_df)
plt.title('Number of Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='winner', data=ipl_df)
plt.title('Number of Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(ipl_df['runs_scored'], bins=8)
plt.title('Distribution of Runs Scored per Match')
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.show()
```
